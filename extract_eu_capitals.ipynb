# Creation of EU capitals dataset

```python
import pandas as pd

capitals = pd.read_csv("datasets/world_capitals.csv")

# Filtering for European capitals
eu_capitals = capitals[capitals["ContinentName"] == "Europe"]

# Selecting relevant columns and renaming them
eu_capitals = eu_capitals[["CapitalName", "CapitalLatitude", "CapitalLongitude"]]
eu_capitals.columns = ["name", "latitude", "longitude"]

# Rounding to the nearest 0.25 degree
eu_capitals["latitude"] = (eu_capitals["latitude"] / 0.25).round() * 0.25
eu_capitals["longitude"] = (eu_capitals["longitude"] / 0.25).round() * 0.25

eu_capitals.head()
```

```python
print(f"Extracted {len(eu_capitals)} European capitals.")
```
